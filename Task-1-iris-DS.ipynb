## Load the Dataset

```python
# Load iris dataset from seaborn
iris = sns.load_dataset('iris')
```

## Data Visualization

### Scatter Plot (Feature Relationship)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')
plt.show()
```

### Histograms (Feature Distributions)

```python
iris.hist(figsize=(10,8))
plt.suptitle('Feature Distributions')
plt.show()
```

### Box Plots (Outlier Detection)

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title('Box Plot of Iris Features')
plt.show()
```

### Final Insights & Observations

1. Dataset is clean (no missing values)
2. Petal features are more informative for species classification
3. Data visualization helps quickly identify patterns
